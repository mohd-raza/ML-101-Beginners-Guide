```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, )
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("My first line graph!!")
plt.show()
```

```python
labels = ['A','B','C']
values = [1,2,3]
plt.bar(labels,values)
plt.xlabel("Labels")
plt.ylabel("Values")
plt.show()
```
